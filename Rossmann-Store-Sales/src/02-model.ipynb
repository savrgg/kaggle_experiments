```python
import pandas as pd
import numpy as np
import xgboost as xgb
from sklearn.model_selection import train_test_split
from itertools import product
import matplotlib.pyplot as plt
import seaborn as sns

def weight_function(y):
    w = np.zeros(y.shape, dtype=float)
    ind = y != 0
    w[ind] = 1. / (y[ind] ** 2)
    return w

def rmspe(yhat, y):
    w = weight_function(y)
    rmspe = np.sqrt(np.mean(w * (y - yhat) ** 2))
    return rmspe

def rmspe_xg(yhat, dtrain):
    y = dtrain.get_label()
    y = np.exp(y) - 1
    yhat = np.exp(yhat) - 1
    w = weight_function(y)
    rmspe = np.sqrt(np.mean(w * (y - yhat) ** 2))
    return "rmspe", rmspe

def build_features(features, data):
    features.extend(['Store', 'CompetitionDistance', 'CompetitionOpenSinceMonth',
                     'CompetitionOpenSinceYear', 'Promo', 'Promo2', 'Promo2SinceWeek',
                     'Promo2SinceYear', 'DayOfWeek', 'SchoolHoliday'])
    
    data['Year'] = data['Date'].apply(lambda x: int(x.split('-')[0]))
    data['Month'] = data['Date'].apply(lambda x: int(x.split('-')[1]))

    for x in ['StoreType', 'Assortment', 'StateHoliday']:
        features.append(x)
        labels = data[x].unique()
        map_labels = dict(zip(labels, range(0, len(labels))))
        data[x] = data[x].map(map_labels)

def train_xgboost_model(train, features, params, num_trees=300):
    X_train, X_test = train_test_split(train, test_size=10000, random_state=42)
    dtrain = xgb.DMatrix(X_train[features], np.log(X_train["Sales"] + 1))
    dvalid = xgb.DMatrix(X_test[features], np.log(X_test["Sales"] + 1))
    watchlist = [(dvalid, 'eval'), (dtrain, 'train')]
    
    evals_result = {}
    gbm = xgb.train(params, dtrain, num_trees,
                    evals=watchlist,
                    evals_result=evals_result,
                    early_stopping_rounds=50,
                    feval=rmspe_xg, verbose_eval=True)
    return gbm, X_test, evals_result

def evaluate_model(gbm, X_test, features):
    train_probs = gbm.predict(xgb.DMatrix(X_test[features]))
    indices = train_probs < 0
    train_probs[indices] = 0
    error = rmspe(np.exp(train_probs) - 1, X_test['Sales'].values)
    print('Error:', error)

def xgb_cross_val(train, features, params, num_boost_round=300):
    dtrain = xgb.DMatrix(train[features], np.log(train["Sales"] + 1))
    cv_results = xgb.cv(params,
                        dtrain,
                        num_boost_round=num_boost_round,
                        nfold=3,
                        feval=rmspe_xg,
                        early_stopping_rounds=50,
                        verbose_eval=True)
    return cv_results['test-rmspe-mean'].min()

def plot_feature_importance(gbm, features):
    plt.figure(figsize=(10, 12))
    xgb.plot_importance(gbm, importance_type='weight', title='Feature Importance', show_values=False)
    plt.show()

def plot_predictions_distribution(gbm, X_test, features):
    train_probs = gbm.predict(xgb.DMatrix(X_test[features]))
    predictions = np.exp(train_probs) - 1
    true_values = X_test['Sales'].values
    
    plt.figure(figsize=(12, 5))
    sns.distplot(predictions, label='Predictions', hist=False)
    sns.distplot(true_values, label='True Values', hist=False)
    plt.legend()
    plt.title('Distribution of Predictions and True Values')
    plt.show()

def plot_residuals(gbm, X_test, features):
    train_probs = gbm.predict(xgb.DMatrix(X_test[features]))
    predictions = np.exp(train_probs) - 1
    residuals = X_test['Sales'].values - predictions
    
    plt.figure(figsize=(12, 5))
    sns.distplot(residuals, hist=False)
    plt.title('Distribution of Residuals')
    plt.show()


def plot_hyperparameters_performance(param_grid, results):
    df = pd.DataFrame(results, columns=['eta', 'max_depth', 'rmspe'])
    df_pivot = df.pivot_table(index="eta", columns="max_depth", values="rmspe", aggfunc=np.mean)
    plt.figure(figsize=(10, 6))
    sns.heatmap(df_pivot, annot=True, cmap="YlGnBu", cbar_kws={'label': 'RMSPE'})
    plt.title('RMSPE for different hyperparameters')
    plt.show()

if __name__ == "__main__":
    train = pd.read_csv("data/train.csv")
    store = pd.read_csv("data/store.csv")

    train = pd.merge(train, store, on='Store')
    train.fillna(0, inplace=True)
    train = train[train["Open"] != 0]
    train = train[train["Sales"] != 0]

    features = []
    build_features(features, train)

    param_grid = {
        "eta": [0.03, 0.3],
        "max_depth": [8, 10]
    }

    best_params = None
    best_rmspe = float("inf")
    results = []

    for params in product(*param_grid.values()):
        test_params = dict(zip(param_grid.keys(), params))
        test_params["objective"] = "reg:linear"
        test_params["silent"] = 1
        test_params["subsample"] = 0.8
        test_params["colsample_bytree"] = 0.7
        test_params["alpha"] = 0
        test_params["lambda"] = 1

        rmspe_score = xgb_cross_val(train, features, test_params)
        results.append((test_params['eta'], test_params['max_depth'], rmspe_score))
        if rmspe_score < best_rmspe:
            best_rmspe = rmspe_score
            best_params = test_params

    print(f"Best parameters: {best_params}")
    print(f"Best RMSPE score: {best_rmspe}")

    plot_hyperparameters_performance(param_grid, results)

    gbm, X_test, evals_result = train_xgboost_model(train, features, best_params)
    evaluate_model(gbm, X_test, features)

    plt.figure(figsize=(12, 5))
    plt.plot(evals_result['train']['rmspe'], label='Train RMSPE')
    plt.plot(evals_result['eval']['rmspe'], label='Validation RMSPE')
    plt.legend()
    plt.xlabel('Boosting Round')
    plt.ylabel('RMSPE')
    plt.title('RMSPE vs. Boosting Round')
    plt.show()

    plot_feature_importance(gbm, features)
    plot_predictions_distribution(gbm, X_test, features)
    plot_residuals(gbm, X_test, features)

```
